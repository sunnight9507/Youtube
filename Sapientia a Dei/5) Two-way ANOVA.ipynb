# 5-1 Two-way ANOVA의 특이점은? - Two-way ANOVA (1): main effect & interaction effect -

- **`Two-way ANOVA`**
    - **독립변수가 2개 + $\alpha$인 ANOVA**
    - 여기서 독립변수는 **Main effect**(주효과)가 있다고 함
        - Two-way ANOVA에서는 Main effect가 2개
        - **`Interaction`**(상호작용 / 교호작용)
            - 한 독립변수의 main effect가 다른 독립변수의 level(=group)에 따라 원래의 linear relationship이 non-linear하게 변하는 경우
            - **한 독립변수의 종속변수에 대한 영향관계가 다른 독립변수의 level(=group)에 따라 변하는 경우**
            - linear relationship(선형 관계)
                - 독립변수와 종속변수의 관계가 선형관계는 전제되어 있다.
            
#### 5-2 Two-way ANOVA 전에 알아야 할 것들

- 그룹이 2개인 경우에도 t-test 이용 가능
    - t-test 대신 One-way ANOVA 이용 할 때 F-value = (t-value)^2
    
- Two-way ANOVA
    - 독립변수는 2개 => F값이 2개(각각의 변수가 유의한지 확인)
    - interaction이 유의한지 아닌지 알아야 한다. => interaction에 대한 F값도 한 개 더 필요
    - **총 3개의 F-value가 필요** => 3개의 BV, 동일한 WV 
- **통계적 가설은 3개가 필요**
    - 첫 번째 main effect에 대한 통계적 가설
        - $H_{01} : \mu_{11} = \mu_{12} = ... = \mu_{1k}$(k는 그룹의 개수)
        - $H_{a1} : \mu_{1i} \neq \mu_{1j}$ for some $i,j$
        - 유의할 경우 사후검정 필요
    - 두 번쨰 main effect에 대한 통계적 가설
        - $H_{02} : \mu_{21} = \mu_{22} = ... = \mu_{2k}$(k는 그룹의 개수)
        - $H_{a2} : \mu_{2i} \neq \mu_{2j}$ for some $i,j$
        - 유의할 경우 사후검정 필요
    - Interaction에 대한 통계적 가설(interaction 유/무)
        - $H_{03}$ : Interaction 존재하지 않다.
        - $H_{a3}$ : Interaction 존재
        - 유의할 경우 사후검정 가능은 하지만 매우 복잡
        - 사후검정 보다는 그래프를 그려보는 것이 더 이해하기 쉬움
        
#### 5-3 Two-way ANOVA의 F-value(F값)을 계산해 보자

- 예시(두 가지 세탁세제의 세탁력을 비교)
    - 두 가지 세제와 세가지 종류의 물 온도로 나누어 실험
    - 각 케이스 별로 4번의 세탁 실험을 함
    - 주의할 점
        - 종속변수는 1개, 연속형
        - 독립변수가 2개
    - WV 계산 
        - (각 value - 각 그룹 평균)^2 / 자유도 : (각 그룹의 실험 회수 - 1) * 각 그룹의 개수
    - BV 계산(전체 평균으로부터 각 그룹의 분산)
        - 첫번쨰, 두번째 main effect 계산 방법
            - (각 실험 갯수 * 그룹 갯수) * (그룹 평균 - 전체 평균)^2 / 자유도 : 그룹의 갯수 - 1
        - interaction 계산 방법
            - 각 그룹간 실험 횟수 * (그룹1, 그룹2의 평균 - 그룹 1 평균 - 그룹 2 평균 + 전체 평균) ^ 2 / 자유도 : (그룹 1 - 1) * (그룹 2 - 1)

#### 5-5 Two-way ANOVA SS의 종류

- SS(sum of squares)
    - 분산을 계산할 때, 분자 부분을 SS
    - 보통 Type 1/2/3로 이야기 하는데, 이는 통계학에 정해진 것이라기 보다는 SAS가 만들어질 때 프로그램 개발자들이 넣은 것
    - One-way ANOVA 또는 그룹내의 샘플 갯수가 동일할 때는 차이가 없음
    - Unbalanced에만 결과가 달라짐

- Type 1 SS
    - 제 1유형 제곱합 또는 순차 제곱합
    - 변수를 한 개씩 순차적으로 추가하면서 제곱합을 계산
    - 변수의 순서에 따라 SS가 변화
    - 특별한 목적에 맞춰서 사용
        - 중요한 변수를 먼저 감안한 뒤에 다른 추가 변수를 테스트하려 할 경우
        - 변동 불가능한 변수를 먼저 고려한 뒤에 변동 가능한 변수를 넣고 테스트
    - 일반적으로 많이 쓰이지는 않음
    
- Type 2 SS
    - 제 2유형 제곱합
    - interaction은 제외하고 계산
    - interaction은 유의하지 않다고 전제 => 유의하지 않을 경우 사용하는 것이 좋음
    
- Type 3 SS
    - 제 3유형 제곱합 또는 수정 제곱합
    - 다른 모든 요인들이 모두 이미 모형에 들어있다는 가정하에 마지막에 새로 추가되는 요인의 변동을 계산한 제곱합
    - 변수의 순서에 따라 SS가 변화하지 않음
    - 일반적으로 사용되는 SS는 Type 3 SS임
    - default 값
