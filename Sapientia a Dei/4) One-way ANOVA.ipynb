#### 4-2 ANOVA 말고 변수 -One-way ANOVA(2)-

- **독립변수(Independent variable)**
    - 논리적 관계에서의 독립을 의미
    - 주로 인과관계를 많이 연구/조사하므로 이때 이 인과관계에서 독립적인 위치는 결과보다는 원인
    - 인과관계에서 원인이 되는 변수가 독립변수
    
- **종속변수(Dependent variable)**
    - 논리적 관계에서의 종속을 의미
    - 주로 인과관계를 많이 연구/조사하므로 이때 이 인과관계에서 독립적인 위치는 원인보다는 결과
    - 인과관계에서 결과가 되는 변수가 종속변수
    
- **통제변수(Control variable)**
    - 기본적으로는 독립변수와 동일하지만 연구/조사의 주된 관심사가 되는 변수가 아닌 경우
    - 통제변수를 한 개도 감안하지 않을 때 Model Misspecification 발생   
    
#### 4-3 ANOVA 전에 알아야 할 것 -One-way ANOVA(3)-

- One-way ANOVA에 사용되는 변수
    - 종속변수 : 연속형 변수만 가능
    - 독립변수 : 이산형/범주형 변수만 가능
        - 독립변수는 단 하나의 변수만 가능
        - ex) 지불방법 - (자동이체, 신용카드, 전자수표, 종이수표)
        -     영상종류 - (폭력영화, 드라마, 공익광고)

#### 4-4 ANOVA의 기본을 이해하자 -One-way ANOVA(4)-

- **수식** : $Y_{ij} = \mu + \tau_j + e_{ij}$
- ANOVA에서 $\mu$는 관심사가 아니다.
- 관심사는  $\tau$
> - $\tau_j$
    - **$\tau$는 독립변수를 의미**
    - **$j$는 그룹을 의미**

- $Y_{ij}$는 종속변수를 의미
    - $j$ : 독립변수의 그룹을 대표하는 문자
    - $i$ : 그 그룹 내의 ID
- $e_{ij}$는 오차
    - $\tau_j$에 의해 설명되지 않는 오차
    - 오차는 random
        
#### 4-5 ANOVA는 F-value이다.

- F-value
    - **두개의 분산의 비율** => **`분산 분석`**
    - **첫번째 분산(Between Variance)** : GM(전체평균)으로부터 각 그룹의 평균사이 분산
        - 이 분산이 크다는 것 : 전체 평균으로부터 각 그룹의 평균값이 멀리 떨어져 있다.
    - **두번째 분산(Within Variance)** : 그룹내의 분산
    - BV가 WB보다 충분히 커야 BV가 통계적으로 크다고 볼 수 있고, 적어도 어느 한 그룹의 평균값이 전체 평균과는 다르다고 할 수 있다.
    - F-value = BV / WB
    
#### 4-6 ANOVA의 F-value을 계산해보자

- 통계적 가설
    - $H_0$ : $\mu_1 =  ...  = \mu_k:$ (k는 그룹의 개수)
    - $H_1$ : $\mu_i \neq \mu_j$ 적어도 한 그룹의 평균은 다르다.
 
- 계산 식
    - F-value = BV / WV
    - BV = (전체평균 - 각 그룹 평균)의 분산 / 그룹 - 1
    - WV = (각 그룹 평균 - 개인)의 분산 / 전체 - 그룹 수
    
#### 4-7 ANOVA의 F-table과 사후검정

- 단순히 One-way ANOVA의 결과만으로는 어떤 그룹이 어떻게 다른지 알 수 없음
- **ANOVA에서 유의하다는 결과가 나오면 자동으로 사후검정을 해야함**
- 사후검정
    - 일종의 여러 다발의 t-test
    - Type-1 error를 발생시키지 않음 
    - 각 그룹의 평균이 다른 그룹의 평균과 같은지 다른지 개별 비교 가능
    - 종류 : Fisher's LSD / Bonferroni / Sheffe / Turkey / Duncan => 대부분 비슷
    
#### 4-8 ANOVA를 실습해보자

- effect size = $\eta^2$ = SSB / (SSB + SSW) = SSB / SST
    - 회귀분석의 $R^2$와 동일한 의미
    - 종속변수의 전체 분산 중 독립변수에 의해 설명된 분산의 비율을 말함
    - **모형의 설명력**이라고도 함
