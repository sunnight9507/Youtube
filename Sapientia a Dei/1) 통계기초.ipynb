#### 1.1 왜 어려울까? - 통계 공부가 어려웠던 이유-
- 통계적 의사결정
    - 확률 : 어떤 사건이 우연히 발생할 확률이 얼마인지
    
#### 1-2 왜 그럴까? - 통계 책 첫 페이지부터 평균과 표준편차가 나온 이유 -
- 대표값 : 평균, 중간값, 최빈값, 표준편차(분산), 구간, 최소값, 최대값
    - 평균
        - 의미 : 자료의 중심값으로 자료의 특성을 대표하는 값
        - 모든 자료로부터 영향을 받음(이상치 영향이 심함)
        - 분산(표준편차)의 계산 등 통계 분석의 대표적인 값으로 사용
    - 분산
        - 의미 : 내가 가진 자료(데이터)가 평균값을 중심을 퍼져 있는 평균적인 거리
        - n이 아닌 n-1인 이유 : 이미 분자에서 평균값으로 한번씩 다 빼주었기 때문
- 평균, 분산이 중요한 이유
    - 계산이 용이
    - 가우스에 따르면, 데이터 불규칙성이 정규분포를 따르고 있으면 최소제곱법이 가장 좋은 추정 방법이고 그 결과 평균값이 가장 좋은 추정값
    - 체비체프에 따르면, 데이터의 불규칙성이 어떠하든 (평균±2X표준편차) 범위 안에는 반드시 전체의 3/4이상의 데이터가 존재함
    
#### 1-3 왜 유의할까? - 도대체 p-값(p-value)는 무엇인가? -
- p-value : probability value(확률값)
    - 어떤 사건이 우연히 발생할 확률
    - ex) p-값이 0.05보다 작다는 것 : 이 사건이 우연히 발생할 확률이 5%보다 작은 것
    - => 우연히 발생한 것이 아니라 인과관계가 있다고 추정
    
#### 1-4 왜 이렇게 복잡할까? - 통계적 가설과 오류는 무엇인가? -
- H0, H1
    - H0 : Null Hypothesis (귀무가설)
        - 아무 영향도 없고 아무 일도 없다
        - 우연히 일어났다고 가정
    - H1 : Alternative Hypothesis (대립가설)
        - 인과관계에 의해 일어났다고 가정

- 오류
    - 1종 오류 : 실제로는 참인데 연구결과가 거짓
    - 2종 오류 : 실제로는 거짓인데 연구결과가 참
    - 1종 오류(a)의 기준이 5%이고, 이 값이 p-값의 기준이 되는 것

#### 1-5 왜 이렇게 생겼을까? - 변수와 데이터 -
- 변수
    - 의미 : 하나의 개념을 대표하는 상징으로서 그것의 특성이 갖는 값이나 강도 또는 크기의 차이를 나타낼 수 있는 것(변하는 숫자)
    - 반대말은 상수(변하지 않는 숫자)
- 변수의 종류
    - 이산형/범주형(Discrete/Categorical)
        - 명목변수/척도(Nominal)
            - 각 범주(속성)간 순위가 없다.
            - 범주에 할당된 값은 의미가 없다.
            - ex) 성별/인종/혈액형 등
        - 순위변수/척도(Ordinal)
            - 각 범주(속성)간 순위가 있다.
            - 순위 사이에 등간성은 없다.
            - ex) 성적/학력/경제수준
    - 연속형(Continuous)
        - 구간변수/척도(Interval)
            - 측정된 범주 사이에 등간성이 존재
            - 할당된 값은 임의 단위로서 비율이나 절대 '0'의 의미가 없다.
            - 덧셈은 가능하지만 곱셈은 안된다.
            - ex) 온도
        - 비율변수/척도(Ratio)
            - 측정된 범주 사이에 등간성이 존재
            - 할당된 값은 임의 단위로서 비율과 절대 '0'의 의미가 있다.
            - 덧셈, 곱셈 모두 가능
            - ex) 키/몸무게/나이/시간
